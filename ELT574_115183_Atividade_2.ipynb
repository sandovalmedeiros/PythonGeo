1 - Ajuste um modelo de SVM de margens rígidas para classificar as amostras linearmente separáveis do problema, considerando as variáveis de entrada "petal length in cm" e "petal width in cm" do banco de dados.  Adicione o ponto [2.5, 1.1] do tipo Iris setosa e explique o que acontece com a margem rígida da SVM.
Para os dados utilizados no anterior, ajuste um modelo de SVM de margens suaves, com os valores C = 1, 10, 100 e 1000.

**Pergunta**

*Comente o que acontece com as margens de cada um dos modelos. Qual se assemelha mais com o modelo de SVM de margens rígidas?*

**Resposta**

Ao ajustar modelos de SVM de margens suaves com diferentes valores de C, podemos observar o comportamento das margens. O valor de C controla o equilíbrio entre o ajuste correto dos dados de treinamento (redução do erro) e a busca por margens maiores.

Em geral, quanto maior o valor de C, menor será a margem, pois o modelo se ajustará mais aos dados de treinamento. Por outro lado, valores menores de C permitem margens maiores e maior tolerância a erros de classificação.

Se observarmos os modelos ajustados com valores de C diferentes, podemos notar que:

Para valores menores de C, como C = 1, as margens são mais amplas, permitindo mais exemplos de treinamento a serem classificados dentro da margem ou até mesmo incorretamente. Esses modelos tendem a ter um maior viés (bias) e uma menor variância.

À medida que o valor de C aumenta, como C = 10, 100 e 1000, as margens são reduzidas e os modelos se ajustam mais aos dados de treinamento. Isso pode resultar em menor viés, mas maior variância, pois o modelo é mais sensível aos dados de treinamento e pode ser mais propenso a superajuste.

Em relação ao modelo de SVM de margens rígidas, o modelo com C muito grande, ou seja, um valor próximo de infinito, se assemelha mais a ele. Isso ocorre porque, nesse caso, o modelo buscará ajustar os dados de treinamento o mais corretamente possível, tentando evitar qualquer erro de classificação, resultando em margens muito estreitas ou até mesmo ausentes. No entanto, é importante destacar que, em certos casos, dependendo da natureza dos dados, um modelo com uma margem mais ampla (C menor) pode ser preferível para evitar o superajuste e ter uma melhor capacidade de generalização para dados não vistos anteriormente.

2 - Ajuste modelos de SVMs não lineares com funções Kernel Polinomial e Gaussiana para os dados do Item 2.

**Pergunta**

*Explique o que acontece com o hiperplano de separação para cada uma das funções Kernel utilizadas.*

**Resposta**

As funções de kernel são usadas para transformar os dados de entrada em um espaço de maior dimensionalidade, onde é mais fácil encontrar um hiperplano de separação entre as classes.

Kernel Polinomial:

O kernel polinomial aplica uma transformação polinomial aos dados de entrada, levando-os a um espaço de dimensionalidade superior.
Com um kernel polinomial, o hiperplano de separação pode se tornar uma curva complexa no espaço de alta dimensionalidade.
A forma exata do hiperplano de separação depende do grau do polinômio utilizado no kernel.
Um grau mais alto de polinômio permite que o hiperplano de separação se torne mais complexo e flexível para se adaptar aos dados.

Kernel Gaussiano (RBF):

O kernel Gaussiano, também conhecido como RBF (Radial Basis Function), mapeia os dados de entrada para um espaço de dimensionalidade infinita.
O hiperplano de separação no espaço transformado é uma superfície não-linear, como uma curva ou uma fronteira complexa.
A forma exata do hiperplano de separação depende dos parâmetros do kernel, como a largura de banda ou a escala do kernel (gamma).
Um valor menor de gamma produz um hiperplano de separação mais suave, enquanto um valor maior de gamma pode levar a um hiperplano de separação mais fragmentado, ajustando-se aos pontos de treinamento mais próximos.

Em resumo, as funções de kernel, como o polinomial e o Gaussiano (RBF), permitem que os modelos SVM ajustem hiperplanos de separação complexos e não lineares nos espaços de alta dimensionalidade resultantes. O hiperplano de separação pode assumir diferentes formas e complexidades, dependendo da função de kernel escolhida e dos parâmetros associados a ela. A escolha da função de kernel apropriada dependerá da natureza dos dados e do problema em questão.

3 - Ajuste um modelo de SVM não linear com função Kernel Gaussiana para os valores de gamma = 0.1, 1 e 10, e valores de C = 1, 10, 100 e 1000.

**Pergunta**

*O que acontece ao aumentar o valor de cada parâmetro da função gaussiana? Como estes aumentos afetam a classificação de amostras do problema?*

**Resposta**

Ao aumentar o valor de cada parâmetro da função gaussiana, que é usada como um kernel no SVM, as seguintes alterações podem ser observadas:

Gamma (γ):

O parâmetro gamma controla a influência de cada exemplo de treinamento na formação do hiperplano de separação.

Aumentar o valor de gamma torna o modelo mais sensível aos pontos de treinamento próximos ao hiperplano de separação.

Com um valor baixo de gamma, o modelo considera uma área mais ampla ao formar o hiperplano, resultando em uma superfície de decisão suave e gradual.
Por outro lado, um valor alto de gamma faz com que o modelo se concentre nos pontos de treinamento mais próximos, resultando em uma superfície de decisão mais fragmentada e ajustada aos pontos de treinamento.
C:

O parâmetro C controla o equilíbrio entre o ajuste correto dos dados de treinamento (redução do erro) e a busca por margens maiores.
Aumentar o valor de C faz com que o modelo SVM tente se ajustar mais aos dados de treinamento.

Um valor alto de C faz com que o modelo busque classificar corretamente a maioria (ou todos) os exemplos de treinamento, resultando em uma superfície de decisão que pode ser ajustada até mesmo a pontos de treinamento mais ruidosos.
Por outro lado, um valor baixo de C permite margens maiores e maior tolerância a erros de classificação, resultando em uma superfície de decisão mais suave e menos propensa a superajuste.

Em relação à classificação das amostras, aumentar o valor de gamma e/ou C pode ter os seguintes efeitos:

Aumentar o valor de gamma pode resultar em uma maior precisão na classificação dos pontos de treinamento próximos, mas pode levar a uma pior generalização para dados não vistos anteriormente (overfitting).

Aumentar o valor de C pode levar a um ajuste mais rigoroso aos dados de treinamento, resultando em um melhor desempenho na classificação dos exemplos de treinamento, mas também aumentando a probabilidade de superajuste.

Portanto, ao aumentar esses parâmetros, é importante considerar a tendência de superajuste e a capacidade de generalização do modelo para dados não vistos anteriormente.

É necessário encontrar um equilíbrio adequado para obter um bom desempenho de classificação em dados novos e desconhecidos.

```python
# Importando as Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

```python
# Carregar o dataset Iris
iris = datasets.load_iris()
X = iris.data[:, 2:4]  # Variáveis de entrada: petal length e petal width
y = iris.target
```

```python
# Ajustar um modelo SVM de margens rígidas
svm_rigid = SVC(kernel='linear', C=1.0)
svm_rigid.fit(X, y)
```

```python
# Adicionar o ponto [2.5, 1.1] do tipo Iris setosa
X = np.vstack([X, [2.5, 1.1]])
y = np.append(y, 0)
```

```python
# Ajustar um novo modelo SVM de margens rígidas com o ponto adicionado
svm_rigid_new = SVC(kernel='linear', C=1.0)
svm_rigid_new.fit(X, y)
```

```python
# Explorar o impacto do ponto adicionado na margem rígida da SVM
margin_diff = svm_rigid_new.decision_function(X[-1].reshape(1, -1)) - svm_rigid_new.decision_function(X[:-1])
print(f"Diferença na margem rígida: {margin_diff}")
```

```python
# Ajustar modelos de SVM de margens suaves com diferentes valores de C
C_values = [1, 10, 100, 1000]
svm_soft_margins = []
for C in C_values:
    svm_soft = SVC(kernel='linear', C=C)
    svm_soft.fit(X, y)
    svm_soft_margins.append(svm_soft)
```

```python
# Comentários sobre as margens dos modelos de SVM de margens suaves
for i, svm_soft in enumerate(svm_soft_margins):
    print(f"Margem para C={C_values[i]}: {1 / np.linalg.norm(svm_soft.coef_)}")
```

```python
# Ajustar modelos de SVM não lineares com diferentes funções de kernel
kernel_models = [
    SVC(kernel='poly', degree=2),
    SVC(kernel='rbf'),
]
```

```python
# Criando a Função que Plota a Decisão
def plot_decisao(model, X, y):
    h = 0.02  # Resolução do grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.xlabel('Comprimento Pétala (cm)')
    plt.ylabel('Largura Pétala (cm)')
    plt.title('Limite')
    plt.show()
```

```python
# Explorar o hiperplano de separação para cada função de kernel
for model in kernel_models:
    model.fit(X, y)
    plot_decisao(model, X, y)
```

```python
# Ajustar um modelo de SVM não linear com função de kernel Gaussiana para diferentes valores de gamma e C
gamma_values = [0.1, 1, 10]
C_values = [1, 10, 100, 1000]
svm_gaussian_models = []
for gamma in gamma_values:
    for C in C_values:
        svm_gaussian = SVC(kernel='rbf', gamma=gamma, C=C)
        svm_gaussian.fit(X, y)
        svm_gaussian_models.append(svm_gaussian)
```

```python
# Comentários sobre o impacto dos valores de gamma e C na classificação das amostras
for svm_gaussian in svm_gaussian_models:
    print(f"Acurácia para gamma={svm_gaussian.gamma}, C={svm_gaussian.C}: {svm_gaussian.score(X, y)}")
```
