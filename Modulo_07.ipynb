```python
# Importando o GeoPandas
import geopandas as gpd
```

```python
# Importando as Geometrias da Biblioteca shapely, caso não tenha instalado faça com o comando PIP
from shapely import Polygon, LineString, Point
```



> **GeoSeries**



```python
 # Criando uma Serie
 series = gpd.GeoSeries(
  
  [   

    Polygon([(0,0), (1,1), (0,1)]),
    Polygon([(10,0),(10,5),(0,0)]),
    Polygon([(0,0),(2,2),(2,0)]),
    LineString([(0,0), (1,1), (0,1)]),
    Point([(0,1)]) 
   
  ]

 )
```

```python
# Exibindo o resultado
series
```

####Observe que o nosso GeoDataFrame vai ter uma coluna chamada: geometry

```python
# Verificando se o objeto é um GeoSeries
type(series)
```
